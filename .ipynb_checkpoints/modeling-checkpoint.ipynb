```python
"""
S&P 500 Stock Price Forecasting Pipeline
Includes: Feature Engineering, ARIMA, SARIMA, Prophet, XGBoost, LightGBM
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Time series models
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
# from prophet import Prophet
# import pmdarima as pm

# ML models
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import xgboost as xgb


# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
df = pd.read_parquet('./cleaned_s&p500.parquet')

stock = 'AAPL'
df_stock = df[df['Name'] == stock].copy()

df_stock['date'] = pd.to_datetime(df_stock['date'])
df_stock = df_stock.sort_values('date')
df_stock.set_index('date', inplace=True)
df_stock = df_stock.asfreq('B')
print(df_stock.head())
print(df_stock.shape)
```

```python
# check stationarity
df_stock['close'].plot();
```

```python
def check_stationarity(timeseries, title='Time Series'):
    print(f'\n{title} - Stationarity Test:')
    print('-' * 50)

    result = adfuller(timeseries.dropna())

    print(f'ADF Statistic: {result[0]:.6f}')
    print(f'p-value: {result[1]:.6f}')

    if result[1] <= 0.05:
        print("Stationary")
    else:
        print("Non-Stationary")

check_stationarity(df_stock['close'], 'Close Price')
```

```python
result = seasonal_decompose(df_stock['close'], model='additive', period=252)
```

```python
result.plot()
```

```python
check_stationarity(df_stock['close'].diff(), 'Close Price')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

plot_acf(series, ax=axes[0], lags=30)
axes[0].set_title("ACF - Close Price")

plot_pacf(series, ax=axes[1], lags=50)
axes[1].set_title("PACF - Close Price")

plt.show()
```

```python
series = df_stock['close'].diff(2).dropna() # 1 diff
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

plot_acf(series, ax=axes[0], lags=30)
axes[0].set_title("ACF - Close Price")

plot_pacf(series, ax=axes[1], lags=50)
axes[1].set_title("PACF - Close Price")

plt.show()
```

```python
# conclusion 
# acf drops sharply after 1 lag , pacf falls slowly SO the appropriate model will be MA
```

```python
df_feat = df_stock.copy()

# Lag features
for lag in [1,2,3,5,7,14,21,30]:
    df_feat[f'close_lag_{lag}'] = df_feat['close'].shift(lag)

# Rolling stats
for w in [5,10,20,50,100]:
    df_feat[f'ma_{w}'] = df_feat['close'].rolling(w).mean()
    df_feat[f'ma_std_{w}'] = df_feat['close'].rolling(w).std()

# Returns
df_feat['returns'] = df_feat['close'].pct_change()

# RSI
delta = df_feat['close'].diff()
gain = delta.clip(lower=0).rolling(14).mean()
loss = -delta.clip(upper=0).rolling(14).mean()
rs = gain / loss
df_feat['rsi'] = 100 - (100 / (1 + rs))

# Time features
df_feat['day_of_week'] = df_feat.index.dayofweek
df_feat['month'] = df_feat.index.month

df_feat.dropna(inplace=True)

print(df_feat.shape)
```

```python
df_stock['log_return'] = np.log(df_stock['close']).diff()
returns = df_stock['log_return'].dropna()
```

```python
from arch import arch_model
model = arch_model(
    returns,
    mean='AR',
    lags=1,
    vol='Garch',
    p=1,
    q=1,
    dist='t'   # important for fat tails
)

res = model.fit(disp='off')
print(res.summary())
```

```python
std_resid = res.std_resid

plt.figure()
plt.hist(std_resid, bins=50)
plt.title("Standardized Residuals")
plt.show()
```

```python
from scipy import stats
import statsmodels.api as sm
# Standardized residuals
std_resid = res.std_resid

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Standardized residuals over time
axes[0,0].plot(std_resid)
axes[0,0].set_title('Standardized Residuals')
axes[0,0].axhline(0, color='red', linestyle='--')

# 2. ACF of standardized residuals
sm.graphics.tsa.plot_acf(std_resid, lags=20, ax=axes[0,1])
axes[0,1].set_title('ACF of Standardized Residuals')

# 3. ACF of squared standardized residuals (check ARCH effects)
sm.graphics.tsa.plot_acf(std_resid**2, lags=20, ax=axes[1,0])
axes[1,0].set_title('ACF of Squared Residuals')

# 4. Q-Q plot vs. normal
stats.probplot(std_resid, dist="norm", plot=axes[1,1])
axes[1,1].set_title('Q-Q Plot')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Example SARIMA(p,d,q)(P,D,Q,252)
model = SARIMAX(
    df_stock['close'],
    order=(1, 1, 1),              # non-seasonal part
    seasonal_order=(1, 1, 1, 252), # yearly seasonality
    enforce_stationarity=False,
    enforce_invertibility=False
)

res = model.fit(disp=False)
```
